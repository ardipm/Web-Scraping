# Convert into dataframe

```html 

<div _ngcontent-web-gramedia-c26="" class="content-wrap"><!----><p _ngcontent-web-gramedia-c26="" class="div-author ng-star-inserted"><!----><span _ngcontent-web-gramedia-c26="" class="list-author ng-star-inserted"> Nobuhiro Watsuki </span></p><div _ngcontent-web-gramedia-c26="" class="list-title">Samurai X Hokkaido Arc vol. 01</div><div _ngcontent-web-gramedia-c26="" class="formats"><!----><div _ngcontent-web-gramedia-c26="" class="ng-star-inserted"><p _ngcontent-web-gramedia-c26="" class="formats-price">Rp 40.000</p><div _ngcontent-web-gramedia-c26="" class="formats-promo"><p _ngcontent-web-gramedia-c26="" class="formats-base-price">Rp 50.000</p></div></div><!----><!----></div><!----></div>

````python
# initiate list 
list_title = []
list_authors = []
list_price = []
list_images = []

# looping page number
for page in range(1, 11):
    # url
    url = (f'https://www.gramedia.com/categories/buku?page={page}').format(page)

    # access web 
    driver.get(url)

    # define height
    height = int(driver.execute_script("return document.body.scrollHeight"))
    for i in range(1, height, 5):
        driver.execute_script(f"window.scrollTo(0, {i});")

    # extract html
    html = driver.page_source

    # parsing html
    soup = BeautifulSoup(html, 'html.parser')

    # collecting boxes
    boxes = soup.find_all('div', {'class':"content-wrap"})

    # access value in box
    for box in boxes:

        try:
            title = box.find('div', {'class':'list-title'})
            list_title.append(title.get_text())
        except:
            list_title.append(None)

        
        try:
            author = box.find('span', {'class':'list-author ng-star-inserted'})
            list_authors.append(author.get_text())
        except:
            list_authors.append(None)
        
        try:
            price = box.find('p', {'class':'formats-price'})
            list_price.append(price.get_text())
        except:
            list_price.append(None)

        try:
            images = soup.find_all('img',{'class':'product-list-img'})
            list_images.append(image['src'])
        except:
            list_images.append(None)


````

````python
# gabungkan kedalam satu dataframe
data = pd.DataFrame({
    'title': list_title,
    'author': list_authors,
    'price': list_price,
    'image': list_images
})

data.info()
````

# Save to .csv

````python
data.to_csv('data_scraped.csv', index=False)
````

# Exploratory Data Analysis

## Data Preparation

````python
df = pd.read_csv('data_scraped.csv')
df.head()
````

````python
df.duplicated().sum()
````

````python
cleaned_df = df.drop_duplicates()
cleaned_df
````

````python
cleaned_df.isna().sum()
````

````python
cleaned_df = cleaned_df.dropna()
cleaned_df
````

````python
cleaned_df.info()
````

````python
columns_to_clean = ['price']
# assign kedalam variable pada column yang ingin dilakukan converting
cleaned_df_copy = cleaned_df.copy(deep=True)

for i in columns_to_clean: 
    cleaned_df_copy[i] = cleaned_df_copy[i].replace('Rp','', regex=True).replace('\.','', regex=True)
    # menggunakan .replace() guna mengubah value pada column yang dilakukan iterasi dengan menggunakan regex (regular expression)
    
cleaned_df_copy.head()
````

````python
cleaned_df = cleaned_df_copy
cleaned_df
````

````python
cleaned_df.dtypes
````

````python
cleaned_df['price'] = cleaned_df['price'].astype('float64')
cleaned_df.info()
````

````python
del cleaned_df['image']
````

````python
cleaned_df
````

````python
# Calculate the average price
avg_price = cleaned_df['price'].mean()
avg_price
````

````python
# Create a plot for price distribution and average price
plt.figure(figsize=(10, 6))

# Plotting the price distribution
sns.histplot(cleaned_df['price'], kde=True, bins=30, color='blue', label='Price Distribution')

# Plotting the average price
plt.axvline(avg_price, color='red', linestyle='--', linewidth=2, label=f'Average Price: Rp{avg_price:.2f}')

# Adding titles and labels
plt.title('Price Distribution and Average Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()

# Display the plot
plt.show()
````
